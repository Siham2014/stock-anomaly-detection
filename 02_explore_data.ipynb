```python
# Cell 1 - Importations et chargement
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration de l'affichage
plt.style.use('seaborn-v0_8')
pd.set_option('display.max_columns', None)

# Chargement des donn√©es
df = pd.read_csv('stock_data.csv')
print("üìä CHARGEMENT R√âUSSI!")
print(f"Shape du dataset: {df.shape}")
```

```python
# Cell 2 - Conversion des dates et tri
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(['Ticker', 'Date'])

print("‚úÖ Dates converties et donn√©es tri√©es")
print(f"Type de la colonne Date: {df['Date'].dtype}")
```

```python
# Cell 3 - Analyse des tickers et p√©riode
print("üìà TICKERS UNIQUES:")
print(df['Ticker'].value_counts())

print(f"\nüìÖ P√âRIODE COUVERTE:")
print(f"Du {df['Date'].min()} au {df['Date'].max()}")

print(f"\nüî¢ NOMBRE DE JOURS PAR TICKER:")
ticker_counts = df['Ticker'].value_counts()
for ticker, count in ticker_counts.items():
    print(f"- {ticker}: {count} jours")

print(f"\nüìÜ NOMBRE TOTAL DE JOURS: {df['Date'].nunique()}")
```

```python
# Cell 4 - Visualisation des prix de cl√¥ture
plt.figure(figsize=(14, 8))

for ticker in df['Ticker'].unique():
    ticker_data = df[df['Ticker'] == ticker]
    plt.plot(ticker_data['Date'], ticker_data['Close'], label=ticker, marker='o', markersize=3, linewidth=2)

plt.title('√âvolution des Prix de Cl√¥ture par Action', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Prix de Cl√¥ture ($)', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("üìà Graphique des prix de cl√¥ture g√©n√©r√©!")
```

```python
# Cell 5 - Analyse des volumes d'√©change
plt.figure(figsize=(14, 8))

for ticker in df['Ticker'].unique():
    ticker_data = df[df['Ticker'] == ticker]
    plt.plot(ticker_data['Date'], ticker_data['Volume'], label=ticker, alpha=0.7, linewidth=2)

plt.title('Volumes d\'√âchange par Action', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Volume', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.yscale('log')  # √âchelle logarithmique pour mieux voir
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("üìä Graphique des volumes g√©n√©r√©!")
```

```python
# Cell 6 - Matrice de corr√©lation
print("üîç CALCUL DE LA MATRICE DE CORR√âLATION...")

# S√©lection des colonnes num√©riques
numeric_cols = df.select_dtypes(include=[np.number]).columns
correlation_matrix = df[numeric_cols].corr()

plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))  # Masque pour le triangle sup√©rieur
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', center=0, 
            fmt='.2f', square=True, cbar_kws={"shrink": .8})
plt.title('Matrice de Corr√©lation des Features Num√©riques', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

print("‚úÖ Matrice de corr√©lation g√©n√©r√©e!")
```

```python
# Cell 7 - Analyse des distributions
print("üìä ANALYSE DES DISTRIBUTIONS...")

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
features_to_plot = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

for idx, feature in enumerate(features_to_plot):
    ax = axes[idx // 3, idx % 3]
    df[feature].hist(bins=50, ax=ax, alpha=0.7, color='skyblue')
    ax.set_title(f'Distribution de {feature}', fontweight='bold')
    ax.set_xlabel(feature)
    ax.set_ylabel('Fr√©quence')
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Cell 8 - Analyse par ticker (boxplots)
print("üì¶ BOXPLOTS PAR TICKER...")

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
price_features = ['Open', 'High', 'Low', 'Close', 'Adj Close']

for idx, feature in enumerate(price_features):
    ax = axes[idx // 3, idx % 3]
    df.boxplot(column=feature, by='Ticker', ax=ax)
    ax.set_title(f'{feature} par Ticker', fontweight='bold')
    ax.set_ylabel(feature)
    ax.tick_params(axis='x', rotation=45)

plt.suptitle('')  # Supprime le titre automatique
plt.tight_layout()
plt.show()
```

```python
# Cell 9 - Calcul des returns et volatilit√©
print("üìà CALCUL DES RETURNS ET VOLATILIT√â...")

returns_data = []

for ticker in df['Ticker'].unique():
    ticker_data = df[df['Ticker'] == ticker].copy()
    ticker_data = ticker_data.sort_values('Date')
    
    # Calcul des returns journaliers
    ticker_data['Daily_Return'] = ticker_data['Close'].pct_change()
    ticker_data['Cumulative_Return'] = (1 + ticker_data['Daily_Return']).cumprod() - 1
    
    # Volatilit√© (√©cart-type des returns sur 20 jours)
    ticker_data['Volatility_20d'] = ticker_data['Daily_Return'].rolling(window=20).std()
    
    returns_data.append(ticker_data)

# Reconstruction
df_returns = pd.concat(returns_data, ignore_index=True)

print("‚úÖ Returns et volatilit√© calcul√©s!")
print(f"Nouvelles colonnes: {[col for col in df_returns.columns if col not in df.columns]}")
```

```python
# Cell 10 - Visualisation des returns cumulatifs
plt.figure(figsize=(14, 8))

for ticker in df_returns['Ticker'].unique():
    ticker_data = df_returns[df_returns['Ticker'] == ticker].dropna()
    plt.plot(ticker_data['Date'], ticker_data['Cumulative_Return'] * 100, 
             label=ticker, linewidth=2.5)

plt.title('Returns Cumulatifs par Action (%)', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Return Cumulatif (%)', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("üìà Graphique des returns cumulatifs g√©n√©r√©!")
```
